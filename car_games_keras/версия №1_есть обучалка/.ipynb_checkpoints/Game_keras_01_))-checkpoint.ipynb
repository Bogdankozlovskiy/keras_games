```python
import pygame
from pygame import*
import numpy as np
from math import pi
import random
from world import *
import os

from collections import deque
from keras.layers.core import Dense,Activation,Flatten,Dropout
from keras.layers import Input
from keras.optimizers import Adam
from keras.models import Model,Sequential

np.random.seed(23)
random.seed(23)

sizeshow=(800,600)
len_step=0.05                          #размер одного шага
m = generate_map(8, 6, 3, 3)          #создаем карту
s=create_car()                        #создаем машинку
bg=pygame.Surface(sizeshow)           #создали фон
car_start=State((m[0][0] + m[0][1]) / 2,rect(len_step, pi*3/4))
car=car_start.copy()
er=np.mean(np.array(m),axis=0)[0]     #кардинаты центра карты
phase0=0
```

```python
class MyGame(object):
    def __init__(self):
        self.len_step=0.025
        self.start_phase=pi*3/4
        self.map=generate_map(8,6,3,3)
        self.start_pos=State((self.map[0][0] + self.map[0][1]) / 2,rect(self.len_step, self.start_phase))
        self.car=create_car()
        self.pointmap=np.mean(np.array(self.map),axis=0)[0]
        self.num_layers=7
    def restart(self):
        self.deq=deque(maxlen=4)
        self.car_position=self.start_pos.copy()
        self.phase = 100
    def step(self,step):
        self.car_position=neuron(self.car_position,step)
        for i in range(4):
            lit=ray(self.car,self.map,self.car_position,self.num_layers)
            self.deq.append(lit)
            self.car_position.position = self.car_position.position + self.car_position.direction
        self.phase,self.exity=carr(self.pointmap,self.car_position,self.phase)
        if is_out_of_map(self.map,self.car_position.position) or self.exity:       #если произошло столкновение с картой
            return -1,np.array(self.deq).reshape(4*self.num_layers)
        else:
            return 1,np.array(self.deq).reshape(4*self.num_layers)
```

```python
from keras.models import Sequential
from keras.layers.core import Activation,Dense,Flatten
from keras.layers.convolutional import Conv2D
from keras.optimizers import Adam
from scipy.misc.pilutil import imresize 
```

```python
def preprocess_images(images):
    if images.shape[0] < 4:
        x_t = images[0]
        x_t = imresize(x_t,(80,80))
        x_t = x_t.astype("float")
        x_t /= 255.0
        s_t = np.stack((x_t,x_t,x_t,x_t),axis=2)
    else:
        xt_list=[]
        for i in range(images.shape[0]):
            x_t = imresize(images[i],(80,80))
            x_t = x_t.astype("float")
            x_t /= 255.0
            xt_list.append(x_t)
        s_t = np.stack((xt_list[0],xt_list[1],xt_list[2],xt_list[3]),axis=2)
    s_t = np.expand_dims(s_t,axis=0)
    return s_t

def get_next_batch(experince,model,num_actions,gamma,batch_size):
    batch_indices = np.random.randint(low=0,high=len(experince),size=batch_size)
    batch = [experince[i] for i in batch_indices]
    X = np.zeros((batch_size,80,80,4))
    Y = np.zeros((batch_size,num_actions))
    for i in range(len(batch)):
        s_t,a_t,r_t,s_tp1,game_over = batch[i]
        X[i] = s_t
        Y[i] = model.predict(s_t)[0]
        Q_sa = np.max(model.predict(s_tp1)[0])
        if game_over:
            Y[i,a_t] = r_t
        else:
            Y[i,a_t] = r_t + gamma * Q_sa
    return X,Y
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=8, strides=4,kernel_initializer="normal",padding="same",input_shape=(80,80,4)))
model.add(Activation("relu"))
model.add(Conv2D(64, kernel_size=4, strides=2,kernel_initializer="normal",padding="same"))
model.add(Activation("relu"))
model.add(Conv2D(64, kernel_size=3, strides=1,kernel_initializer="normal",padding="same"))
model.add(Activation("relu"))
model.add(Flatten())
model.add(Dense(512,kernel_initializer="normal"))
model.add(Activation("relu"))
model.add(Dense(3,kernel_initializer="normal"))
model.compile(optimizer=Adam(lr=1e-6),loss="mse")
```

```python
NUM_ACTIONS = 3
GAMMA = 0.99
INITIAL_EPSILON = 0.1
FINAL_EPSILON = 0.0001
MEMORY_SIZE = 7000
NUM_EPOCHS_OBSERVE = 100
NUM_EPOCHS_TRAIN = 5000

BATCH_SIZE = 32
NUM_EPOCHS = NUM_EPOCHS_OBSERVE + NUM_EPOCHS_TRAIN

game = MyWrappedGame()
experince = collections.deque(maxlen=MEMORY_SIZE)

num_games,num_wins=0,0
epsilon = INITIAL_EPSILON
```

```python
for e in range(NUM_EPOCHS):
    game.reset()
    loss = 0.0
    a_0=1
    x_t,r_0,game_over = game.step(a_0)
    s_t = preprocess_image(x_t)
    game_over = False
    while not game_over:
        s_tml = s_t
        #следующее действие
        if e <= NUM_EPOCHS_OBSERVE:
            a_t = np.random.randint(low=0,high=NUM_ACTIONS,size=1)[0]
        else:
            if np.random.rand() <= epsilon:
                a_t = np.random.randint(low=0,high=NUM_ACTIONS,size=1)[0]
            else:
                q = model.predict(s_t)[0]
                a_t = np.argmax(q)
        # применить действие к игре
        x_t, r_t, game_over=game.step(a_t)
        s_t = preprocess_images(x_t)
        #если вознагрождение положительно, увеличить num_wins
        if r_t == 1:
            num_wins += 1
        #сохраним опыт
        experince.append((s_tml, a_t, r_t, s_t, game_over))
        if e > NUM_EPOCHS_OBSERVE:
            X,Y = get_next_batch(experince,model,NUM_ACTIONS,GAMMA,BATCH_SIZE)
            loss += model.train_on_batch(X,Y)
            if epsilon > FINAL_EPSILON:
                epsilon -= (INITIAL_EPSILON - FINAL_EPSILON) / NUM_EPOCHS
    print("Epoch {:04d}/{:d} | Loss {:.5f} | Win Count {:d}".format(e + 1,NUM_EPOCHS,loss,num_wins))
```
